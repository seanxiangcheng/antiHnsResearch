```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
##### Internal Energy Plots #####
h3ergf = "HN3_AFM_IE.csv"
h5ergf = "HN5_AFM_IE.csv"
hpergf = "HNNP_AFM_IE.csv"
h6ergf = "HN6_AFM_IE.csv"
h3emcf = "H3AF_E_116"
h5emcf = "H5AF_E_116"
hpemcf = "HPAF_E_116"
h6emcf = "H6AF1D_E_k15_r0"
```

```python
# HN3 IE RG plot
h3erg = np.loadtxt(h3ergf, delimiter=',')
d = h3erg
muinv = d[1:,0]
T=d[1:, 1]
p_cols = [8, 16, 32, 64, 128]
p_cols = np.array(p_cols)-4
ie = d[1:,p_cols]
labels = ["k="+str(i+4) for i in p_cols]
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
for y, lab in zip(np.transpose(ie), labels):
    plt.plot(muinv, y, label = lab, linewidth=2)
plt.xlabel("$1/\mu$", fontsize = 16)
plt.ylabel("$<e>$", fontsize = 16)
plt.xlim([0, 1])
plt.ylim([np.min(ie), np.max(ie)])
plt.legend(loc = 'lower right', fontsize=12)

plt.subplot(2,2,2)
for y, lab in zip(np.transpose(ie), labels):
    plt.plot(muinv, y, label = lab, linewidth=2)
plt.xlabel("$1/\mu$", fontsize = 16)
#plt.ylabel("$<e>$", fontsize = 16)
#plt.xlim([0, 1])
#plt.ylim([np.min(ie), np.max(ie)])
plt.xlim([0, 0.1])
plt.ylim([-1, -0.82])
plt.legend(loc = 'lower right', fontsize=12)

plt.subplot(2,2,3)
for y, lab in zip(np.transpose(ie), labels):
    plt.plot(T, y, label = lab, linewidth=2)
plt.xlabel("$T$", fontsize = 16)
plt.ylabel("$<e>$", fontsize = 16)
plt.xlim([0, 12])
plt.ylim([np.min(ie), np.max(ie)])
#plt.xlim([0, 0.1])
plt.ylim([-1, -0.1])
plt.legend(loc = 'lower right', fontsize=12)

plt.subplot(2,2,4)
for y, lab in zip(np.transpose(ie), labels):
    plt.plot(T, y, label = lab, linewidth=2)
plt.xlabel("$T$", fontsize = 16)
#plt.ylabel("$<e>$", fontsize = 16)
plt.xlim([0, 0.8])
#plt.xlim([0, 0.1])
plt.ylim([-1, -0.85])
plt.legend(loc = 'upper left', fontsize=12)
plt.savefig("HN3_IE_rg.pdf", bbox_inches='tight')

# HN3 RG/MC plot
plt.figure(figsize=(5,4))
plt.plot(T,ie[:,2]+1.,'k-', linewidth=2, label="RG")
d = np.loadtxt(h3emcf)
plt.plot(d[:,0], d[:,1:]+1.0,":", linewidth=1)
plt.xlim([0, 0.98])
plt.ylim([0, 0.2])
plt.xlabel("$T$", fontsize=16)
plt.ylabel("<e> - GS", fontsize=16)
plt.title("HN3 IE of RG & MC, $k=16$", fontsize=16)
plt.plot([],[],'b:',label="MC", linewidth=3)
plt.legend(loc="lower right", fontsize=14)
plt.savefig("HN3_IE_rgmc.pdf", bbox_inches='tight')
```
