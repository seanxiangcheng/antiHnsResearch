## Plot of K vs mu and J vs T

```python
# HNNP J vs T
nt = 50
Tmu3 = 1.820478
Tmin = 0.02
Tmax = 3.2
minv_min = 0.01
minv_max = 1./np.exp(2./Tmax)
n = 2000
l = 120
y = 0



minvs = np.linspace(minv_min, 1.0/3.0, nt)
hp = hnrg.hp6()
plt.figure(figsize=(5, 8))
plt.subplot(211)
for i in xrange(minvs.size):
    minv = minvs[i]
    m = 1.0/minv
    ks = np.unique(hp.rg_n_k(m, y, n, l))
    plt.semilogy(minv*np.ones((ks.size,)), ks, 'ko', markersize = 1)
minvs2 = np.linspace(1.0/3.0, minv_max, 10)
plt.semilogy(minvs2, [1.0]*minvs2.size, 'k-', linewidth=2)
plt.xlim([0.0, minv_max])
plt.xlabel("$1/\mu$", fontsize=16)
plt.ylabel("$\kappa$", fontsize=16)

ts = np.linspace(Tmin, Tmu3, num = nt)
ms = np.exp(2.0/ts)
hp = hnrg.hp6()
plt.subplot(212)
for i in xrange(ms.size):
    m =ms[i]
    t = ts[i]
    js = -np.log(np.unique(hp.rg_n_k(m, y, n, l)))*1.0/4.0
    plt.plot(t*np.ones((js.size,)), js, 'ko', markersize = 1)
ts2 = np.linspace(Tmu3, Tmax, 10)
plt.plot(ts2, [0.0]*ts2.size, 'k-', linewidth=2)
plt.xlim([0.0, Tmax])
plt.ylim([-25, 25])
plt.xlabel("$T$", fontsize=16)
plt.ylabel("$J$", fontsize=16)
plt.savefig("HNNP_RG_Kvsmu_JvsT.pdf")
```

## Plot of boundary of the chaotic part

```python
# HNNP J vs Tc Power law fitting
nt = 100
Tc = 1.8204784532536746
Tmin = 1.68
Tmax = Tc
minv_min = 1./np.exp(2./Tmin)
minv_max = 1./np.exp(2./Tmax)
n = 20000
l = 1000
y = 0
ts = np.linspace(Tmin, Tmax, num = nt)
ms = np.exp(2.0/ts)

hp = hnrg.hp6()

maxminJ = np.zeros((ms.size, 2))
hp = hnrg.hp6()
for i in xrange(ms.size):
    m =ms[i]
    t = ts[i]
    js = -np.log(np.unique(hp.rg_n_k(m, y, n, l)))*1.0/4.0
    maxminJ[i, :] = [max(js), min(js)]
plt.figure(figsize=(6,6))
plt.plot(ts, maxminJ[:, 0], 'k.', markersize = 4)
plt.plot(ts, maxminJ[:, 1], 'r.', markersize = 4)

plt.xlabel("$T$", fontsize=16)
plt.ylabel("$J$", fontsize=16)

# fitting
xs = np.sqrt(Tc-ts)
y1s = maxminJ[:, 0]
y2s = maxminJ[:, 1]

slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(xs, y1s)
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(xs, y2s)

plt.plot(ts, slope1*xs + intercept1, 'g--', linewidth=3)
plt.plot(ts, slope2*xs + intercept2, 'g--', linewidth=3)
plt.xlim([Tmin, Tmax*1.01])
plt.ylim([-0.33, 0.33])
#plt.savefig("JT_fitting_sqrt_deltaT.pdf")
print "Slope          Intercept"
print slope1, intercept1
print slope2, intercept2
```

```python
# HNNP J vs T

def Kmy(m=2.,y=1.0):
    value = (-2*m + m**y + m**(1+y) + m**(y/2.)*(1+m) * np.sqrt(4.*( m-1.)*m + m**y )) / (2*m**2)
    return value


def plot_Kmuy(y=1.0):
    nt = 50
    Tmin = 0.04
    Tmax = 3.2
    minv_min = 0.01
    minv_max = 1.0/np.exp(2./Tmax)
    n = 20000
    l = 120
    minvs = np.linspace(minv_min, minv_max, nt)
    hp = hnrg.hp6()
    #plt.figure(figsize=(5, 8))
    #plt.subplot(211)
    for i in xrange(minvs.size):
        minv = minvs[i]
        m = 1.0/minv
        ks = np.unique(hp.rg_n_k(m, y, n, l))
        plt.semilogy(minv*np.ones((ks.size,)), ks, 'ko', markersize = 1.5)
    kaps2=Kmy(1./minvs, y)
    plt.semilogy(minvs, kaps2, 'r--')
    plt.xlim([0.0, minv_max])
    #plt.xlabel("$\mu$", fontsize=16)
    plt.ylabel("$\kappa$", fontsize=16)
    plt.title("y="+str(y), fontsize=16)


def plot_Jmuy(y=1.0):
    nt = 50
    Tmin = 0.04
    Tmax = 3.2
    minv_min = 0.01
    minv_max = 1.0/np.exp(2./Tmax)
    n = 20000
    l = 120
    ts = np.linspace(Tmin, Tmax, num = nt)
    ms = np.exp(2.0/ts)
    hp = hnrg.hp6()
    #plt.subplot(212)
    for i in xrange(ms.size):
        m =ms[i]
        t = ts[i]
        #print hp.rg_n_k(m, y, n, l)
        js = -np.log(np.unique(hp.rg_n_k(m, y, n, l)))*1.0*t/4.0
        plt.plot(t*np.ones((js.size,)), js, 'ko', markersize = 1.5)
    kaps2 = Kmy(ms, y)
    Js2 = -np.log(kaps2)*1.0*ts/4.0
    plt.plot(ts, Js2, 'r--')
    plt.xlim([0.0, Tmax])
    #plt.ylim([-5, 5])
    #plt.xlabel("$T$", fontsize=16)
    plt.ylabel("$J$", fontsize=16)
    plt.title("y="+str(y), fontsize=16)
#plt.savefig("HNNP_RG_Kvsmu_JvsT.pdf")


plt.figure(figsize=(10,18))

y=2.0
plt.subplot(521)
plot_Kmuy(y)
plt.subplot(522)
plot_Jmuy(y)

y=1.0
plt.subplot(523)
plot_Kmuy(y)
plt.subplot(524)
plot_Jmuy(y)

y=0.
plt.subplot(525)
plot_Kmuy(y)
plt.subplot(526)
plot_Jmuy(y)

y=-1.0
plt.subplot(527)
plot_Kmuy(y)
plt.subplot(528)
plot_Jmuy(y)

y=-2.0
plt.subplot(529)
plot_Kmuy(y)
plt.xlabel("$1/\mu$", fontsize=16)
plt.subplot(5,2,10)
plot_Jmuy(y)
plt.xlabel("$T$", fontsize=16)

#plt.savefig("HNP6_KvsMu_JvsT_ys.pdf")
#plt.savefig("HNP6_KvsMu_JvsT_ys.eps")
```

## HN3 and HN5: analytical solutions

```python
def KJyHN35(m=2.0, y=1.0):
    value = 0.5 * ( -2.*m + m**y + m**(1+y) + np.sqrt( 4.*(-1.+m**y+m**(1.+y)) + (-2.*m+m**y+m**(1.+y))**2 )  )
    return value

def plot_KHN35(y=1.0, sty='-b', lw=2):
    nt = 200
    Tmin = 0.04
    Tmax = 100.2
    minv_min = 0.01
    minv_max = 1.0/np.exp(2./Tmax)
    n = 10000
    l = 120
    minvs = np.linspace(minv_min, minv_max, nt)
    kaps = KJyHN35(1./minvs, y)
    if np.min(kaps)>=0:
        plt.plot(minvs, kaps, sty, linewidth=2, label='y=%.1f' % y)
    else:
        indG0 = kaps>=0.0
        plt.plot(minvs[indG0], kaps[indG0], sty, linewidth=2, label='y=%.1f' % y)
        indG0 = np.logical_not(indG0)
        plt.plot(minvs[indG0], kaps[indG0], sty[0]+'--', linewidth=2)

    plt.xlim([0.0, minv_max])
    plt.ylabel("$\kappa^*$", fontsize=20)
    
def plot_JHN35(y=1.0, sty='-b', lw=2):
    nt = 200
    Tmin = 0.04
    Tmax = 10.2
    minv_min = 0.01
    minv_max = 1.0/np.exp(2./Tmax)
    n = 10000
    l = 120
    ts = np.linspace(Tmin, Tmax, nt)
    ms = np.exp(2.0/ts)
    kaps = KJyHN35(ms, y)
    if np.min(kaps)>1e-15:
        Js = -np.log( kaps )*1.0*ts/4.0
        plt.plot(ts, Js, sty, linewidth=2, label='y=%.1f' % y)
    else:
        indG0 = kaps>1e-15
        Js = -np.log( kaps[indG0] )*1.0*ts[indG0]/4.0
        plt.plot(ts[indG0], Js, sty, linewidth=2, label='y=%.1f' % y)
    plt.xlim([Tmin, Tmax])
    plt.ylabel("$J^*$", fontsize=18)

stys = ['b-', 'r-', 'g-', 'c-', 'k-']
plt.figure(figsize=(8, 4))
plt.subplot(121)
plot_KHN35(y=2.0, sty=stys[0])   
plot_KHN35(y=1.0, sty=stys[1])   
plot_KHN35(y=0.0, sty=stys[2])   
plot_KHN35(y=-1.0, sty=stys[3])   
plot_KHN35(y=-2.0, sty=stys[4])   
plt.axhline(y=0.0, xmin=0, xmax=1, linewidth=2, color = 'y', ls='--')
plt.ylim([-0.2,4])
plt.xlabel("$1/\mu$", fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend(loc='upper left', fontsize=12)

plt.subplot(122)
plot_JHN35(y=2.0, sty=stys[0])   
plot_JHN35(y=1.0, sty=stys[1])   
plot_JHN35(y=0.0, sty=stys[2])   
plot_JHN35(y=-1.0, sty=stys[3])   
plot_JHN35(y=-2.0, sty=stys[4]) 
plt.ylim([-2,6])
plt.xlabel("$T$", fontsize=18)
plt.legend(loc='upper left', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=14)

plt.tight_layout()
plt.savefig("HN35_KJvsTys_the.pdf")
```
